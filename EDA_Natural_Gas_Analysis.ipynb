```
# importando as bibliotecas necessarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# importando o dataframe
df = pd.read_csv('/content/data.csv')
df.head()
```

```
# Resumo da analise estatistica
df.describe()
```

```
# Obter uma visão geral do DataFrame
df.info()
```

```
# visualizando dados das colunas mês e ano
print(df['month'].unique())
print(df['year'].unique())
```

```
# Cria um gráfico de linha da variável 'value' em função da variável 'month' no DataFrame 'df', usando o marcador '^'.
sns.lineplot(data=df,x='month',y='value',marker='^')
plt.show()
```

```
# Calcula o consumo mensal agrupando por 'mês' somando a coluna 'valor' para cada grupo e redefinindo o índice para criar um novo DataFrame chamado 'consumo_mensal'
monthly_consumption = df.groupby('month')['value'].sum().reset_index()
monthly_consumption
```

```
# Cria um gráfico de linha da variável 'value' em função da variável 'month' no DataFrame 'monthly_consumption'
sns.lineplot(data=monthly_consumption,x='month',y='value',marker='o',color='red')
```

```
# Calcula o consumo anual agrupando o DataFrame 'df' por 'ano', somando a coluna 'valor' para cada grupo e redefinindo o índice para criar um novo DataFrame chamado 'year_consumption'.
year_consumption = df.groupby('year')['value'].sum().reset_index()
year_consumption
```

```
#  Cria um gráfico de linha da variável 'value' em função da variável 'year'
sns.lineplot(data=year_consumption,x='year',y='value')
```
